# Loss and Accuracy in Train

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
import tensorflow.keras.models as tf

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (18, 6)
plt.rcParams["legend.facecolor"] = "white"
plt.rcParams["legend.borderpad"] = 1
pallete_colors = cm.get_cmap('Set3', 16)
```

```python
# Load data
dir_stats = './models/stats/'

cnn_ds = pd.read_csv(dir_stats+"model-cnn-deepsound.tsv", sep="\t")
cnn_lstm = pd.read_csv(dir_stats+"model-cnn-lstm-notescaled.tsv", sep="\t")
cnn_sp = pd.read_csv(dir_stats+"model-cnn-spotify.tsv", sep="\t")
```

### DeepSound Model

```python
plt.subplot(1, 2, 1)
plt.plot(cnn_ds['acc'])
plt.plot(cnn_ds['val_acc'])
plt.title('Deepsound Accuracy', pad=20)
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')

plt.subplot(1, 2, 2)
plt.plot(cnn_ds['loss'])
plt.plot(cnn_ds['val_loss'])
plt.title('Deepsound Loss', pad=20)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='best')
plt.savefig(dir_stats+"/graphs/ds.png");
```

### Spotify Model

```python
plt.subplot(1, 2, 1)
plt.plot(cnn_sp['acc'])
plt.plot(cnn_sp['val_acc'])
plt.title('Spotify Accuracy', pad=20)
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')

plt.subplot(1, 2, 2)
plt.plot(cnn_sp['loss'])
plt.plot(cnn_sp['val_loss'])
plt.title('Spotify Loss', pad=20)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='best')
plt.savefig(dir_stats+"/graphs/sp.png");
```

### CNN - LSTM

```python
plt.subplot(1, 2, 1)
plt.plot(cnn_lstm['acc'])
plt.plot(cnn_lstm['val_acc'])
plt.title('CNN-LSTM Accuracy', pad=20)
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')

plt.subplot(1, 2, 2)
plt.plot(cnn_lstm['loss'])
plt.plot(cnn_lstm['val_loss'])
plt.title('CNN-LSTM Loss', pad=20)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='best')
plt.savefig(dir_stats+"/graphs/lstm.png");
```
